```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka iz CSV datoteke
df = pd.read_csv(r'C:\Users\HP PC\Desktop\socmediainfluence.csv')

# Filtriranje podataka za pitanje "What social platform has influenced your online shopping most?"
platform_influence = df[df['Question'] == 'What social platform has influenced your online shopping most?']

# Sumiranje frekvencije odgovora po društvenim platformama
platform_counts = platform_influence.groupby('Answer')['Count'].sum()

# Prikaz bar plot-a za frekvenciju odgovora na temelju društvenih platformi
plt.figure(figsize=(10, 6))

# Kreiranje palete boja
palette = sns.color_palette("husl", len(platform_counts))

plt.bar(platform_counts.index, platform_counts.values, color=palette)  # Dodjela boja iz palete
plt.xlabel('Social platform', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Influence on online shopping by social platform', fontsize=14)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

<div style="font-size: 20px;">
Prema nalazima ankete, većina ispitanika je istaknula da Instagram ima najveći utjecaj na njihovu online kupovinu, vjerojatno zbog svoje vizualne prirode i široke prisutnosti brendova i proizvoda na ovoj platformi.

Grafički prikaz odgovora na pitanje: Koga najviše volite pratiti online?

```python
# Import potrebnih biblioteka
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje podataka iz CSV datoteke
df = pd.read_csv(r'C:\Users\HP PC\Desktop\socmediainfluence.csv')

# Filtriranje podataka za određeno pitanje
df_filtered = df[df['Question'] == 'Who are your favorite people to follow online?']

# Kreiranje jedinstvenih oznaka za legendu
unique_answers = df_filtered['Answer'].unique()

# Kreiranje popisa podataka i odgovarajućih oznaka za kružni grafikon
data = []
labels = []
for answer in unique_answers:
    answer_data = df_filtered[df_filtered['Answer'] == answer]['Count'].sum()
    data.append(answer_data)
    labels.append(answer)

# Prikazivanje podataka pomoću kružnog grafikona
plt.figure(figsize=(8, 8))
plt.pie(data, labels=labels, autopct='%1.1f%%', colors=plt.cm.tab20.colors, textprops={'fontsize': 12})
plt.title('Who are your favorite people to follow online?', fontsize=16)
plt.legend(loc='lower right', fontsize=10)
plt.tight_layout()
plt.show()
```

<div style="font-size: 20px;">
Prema rezultatima ankete, 61.9% ispitanika preferira pratiti svoje prijatelje na društvenim mrežama. Samo 4.4% njih izjavilo je da preferira pratiti poznate osobe, dok 13.9% ispitanika voli pratiti osobe koje se bave zanimljivim aktivnostima ili hobijima. Nadalje, 19.8% ispitanika izjavilo je da najradije prati humoristične profile ili profile koji su izrazito zabavni. Ovi podaci pružaju uvid u preferencije korisnika na društvenim mrežama i njihove različite motive za praćenje određenih profila.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Učitavanje podataka iz CSV datoteke
df = pd.read_csv(r'C:\Users\HP PC\Desktop\socmediainfluence.csv')

# Filtriranje podataka samo za Male i Female voters
male_voters = df[df['Segment Description'] == 'Male voters']
female_voters = df[df['Segment Description'] == 'Female voters']

# Izračun broja muških i ženskih glasova
male_count = male_voters['Count'].sum()
female_count = female_voters['Count'].sum()

# Prikaz podataka u obliku pie chart-a
plt.figure(figsize=(8, 8))
plt.pie([male_count, female_count], labels=['Male Voters', 'Female Voters'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Male vs Female Voters')
plt.show()
```

<div style="font-size: 20px;">
Analizirali smo zastupljenost prema spolu te uočili da je 63 posto ispitanika muškog spola, dok je ostatak ispitanika bio ženskog spola.
</div>


```python
# Filtriranje podataka za zaposlene, nezaposlene i one koji rade na nepuno radno vrijeme
zaposleni = df[df['Segment Description'].str.contains('Yes, full-time')]
nezaposleni = df[df['Segment Description'].str.contains("No, but I'm searching for one")]
zaposleni_nepuno_vrijeme = df[df['Segment Description'].str.contains('Yes, part-time')]

# Izračun broja ispitanika koji su zaposleni, nezaposleni i rade na nepuno radno vrijeme
broj_zaposlenih = zaposleni['Count'].sum()
broj_nezaposlenih = nezaposleni['Count'].sum()
broj_nepuno_vrijeme = zaposleni_nepuno_vrijeme['Count'].sum()

# Prikaz podataka u obliku pie chart-a
plt.figure(figsize=(8, 8))
plt.pie([broj_zaposlenih, broj_nezaposlenih, broj_nepuno_vrijeme], labels=['Zaposleni', 'Nezaposleni', 'Nepuno radno vrijeme'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title('Zaposlenost i utjecaj društvenih platformi na online kupovinu')
plt.show()
```

<div style="font-size: 20px;">
Također smo istraživali zaposlenost ispitanika i primijetili da većina pripada skupini studenata. Među studentima, najviše njih radi na pola radnog vremena, drugi najčešći status je nezaposlenost, dok je najmanji broj ispitanika zaposlen u punom radnom vremenu.

**<div style="font-size: 25px;">Analiza stvarnih i predviđenih vrijednosti</div>**


<div style="font-size: 20px;">
Koristeći navedeni Python kod, izvršili smo predikciju budućih vrijednosti socijalnog utjecaja. Korišten je Random Forest Regressor, algoritam strojnog učenja koji je treniran na povijesnim podacima o socijalnom utjecaju. Podaci su pripremljeni i podijeljeni na skupove za treniranje i testiranje kako bi se model evaluirao na neviđenim podacima. Nakon treniranja, model je korišten za predikciju budućih vrijednosti socijalnog utjecaja, a rezultati su evaluirani koristeći srednju kvadratnu pogrešku kao mjeru performansi modela.
</div>


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Učitavanje podataka
df = pd.read_csv(r'C:\Users\HP PC\Desktop\socmediainfluence.csv')

# Odabir značajki i ciljne varijable
features = ['Segment Type', 'Segment Description', 'Answer', 'Count']  
target = 'Percentage'  

# Pretvaranje kategoričkih značajki u numeričke koristeći one-hot encoding
df_encoded = pd.get_dummies(df[features])

# Razdvajanje podataka na skup za treniranje i testiranje
X_train, X_test, y_train, y_test = train_test_split(df_encoded, df[target], test_size=0.2, random_state=42)

# Inicijalizacija i treniranje modela
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predikcija na skupu za testiranje
y_pred = model.predict(X_test)

# Evaluacija modela
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

<div style="font-size: 20px;">
Što je vrijednost MSE bliža 0, to bolje. Ovaj rezultat ukazuje na to da su predviđanja modela vrlo bliska stvarnim vrijednostima. U ovom slučaju, niska vrijednost MSE sugerira da je model dobro prilagođen podacima i da je njegova prediktivna sposobnost visoka.
</div>


**<div style="font-size: 25px;">Grafički prikaz predikcije za budućnost:</div>**


```python
import matplotlib.pyplot as plt

# Plot stvarnih i predviđenih vrijednosti
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')
plt.xlabel('Stvarne vrijednosti')
plt.ylabel('Predviđene vrijednosti')
plt.title('Stvarne vs. Predviđene vrijednosti')
plt.grid(True)
plt.show()
```

<div style="font-size: 20px;">
Stvarne i predviđene vrijednosti su blizu crvene linije na grafu, to obično znači da je model dobro generaliziran i uspješno predviđa vrijednosti na temelju dostupnih podataka. Može se uočiti da su predviđanja jako slična stvarnim i trenutnim vrijednostima.
